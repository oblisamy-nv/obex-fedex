```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
wiki_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(wiki_url,'lxml')
print (soup.prettify())
```

```python
table = soup.find('div', class_="mw-parser-output")
table = table.find('table', class_="wikitable sortable")
#print (table)
```

```python
headers=[]
for header in table.find_all('th'):
    headers.append(header.text)
headers
toronto_data = pd.DataFrame(columns = headers)
toronto_data
```

```python
i=0
for rows in table.find_all('tr'):
    children = rows.findChildren(recursive=False)
    #print(children)
    row_text = []
    for content in children:
        row_text.append(content.text)
        #print(row_text)
    print(row_text)
    toronto_data.loc[i]=(row_text)
    i=i+1
toronto_data
```
